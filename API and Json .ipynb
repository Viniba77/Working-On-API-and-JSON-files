# Importing multiple records of movies based on id 

```python
movie_id = [0, 299534, 19995, 140607, 299536, 597, 135397,
            420818, 24428, 168259, 99861, 284054, 12445,
            181808, 330457, 351286, 109445, 321612, 260513]
```

```python
basic_url = 'https://api.themoviedb.org/3/movie/{}?{}'
```

```python
json_list=[]
for i in movie_id:
    url =basic_url.format(i,api_key)
    r = requests.get(url)
    if r.status_code !=200:
        continue
    else:
        data = r.json()
        json_list.append(data)
df =pd.DataFrame(json_list)
        
```

```python
requests.get(basic_url.format(0,api_key))
```

```python
df
```

```python
df = df.loc[:,['title','id','revenue','genres','belongs_to_collection','runtime']].sort_values(by='revenue',ascending = False)
df
```

```python
import json
df.to_json('movies.json',orient = 'records')
```
